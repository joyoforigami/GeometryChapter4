```python
from PIL import Image

i = Image.open('/content/'+list(a.keys())[0])

pixels = i.load() 
width, height = i.size


all_pixels = []

for j in range(width):
    for k in range(height):
        cpixel = pixels[j, k]
        all_pixels.append(int(round(sum(cpixel) / float(len(cpixel)))))

pixels = np.array(all_pixels)
pixels = np.round(pixels/36.57)
pixels = np.vectorize(int)(pixels)
for i in range(8):
  print(i,':',list(pixels).count(i))

k=10
pixels = list(pixels)
for i in range(7):
  img = Image.open(str(i)+'.png')
  xx,yy=img.size
  img=img.resize((int(xx/k),int(yy/k)))
  img.save(str(i)+'add.png')

Image.new('RGBA', (1,1), color = (255,255,255)).save('7add.png')
tiles=[Image.open(f'{x}add.png') for x in range(8)]

output = Image.new('RGBA', (width*20,height*20), color = (255,255,255))

def csh(s):
  global pixels
  return sum([pixels.count(int(i)) for i in s])
composition = '''
T
3
AT
3AT
Aa3
Aa3T
Aa3Tå@'''.split('\n')[::-1]


c3 = 0
cT = 0
cA = 0
cAf = 0
cAa = 0
cAb = 0
cc = 0
step = 10

idx = []

with open('instruction.txt', 'w+') as f:
  f.write(f"1. Shift the triangle down 3 spaces and right 8 spaces.\n")
  f.write(f"2. Make 3 copies of the A. \n")
  f.write(f"3. Reflect the first copy over the x-axis. This will make polygon D'E'F'G'H'I'J'K'\n")
  f.write(f"4. Rotate the second copy centered at the origin by 90 degrees. This will make polygon D''E''F''G''H''I''J''K''\n")
  f.write(f"5. Rotate the third copy centered at the origin by 270 degrees. This will make polygon D'''E'''F'''G'''H'''I'''J'''K'''\n")
  f.write(f"6. Shift all of the A's down 3 spaces and right 3 spaces.\n")
  f.write(f"7. Shift the 3 down 3 spaces and left 6 spaces.\n")
  f.write(f"8. Dilate entire image by 0.08.\n")
  f.write(f"9. Make {csh('64310')} copies of polygon ABC (triangle), {csh('43210')} copies of polygon DEFGHIJK (letter A), {csh('210')} copies D'E'F'G'H'I'J'K' (flipped A), {csh('0')} copies of D''E''F''G''H''I''J''K'', {csh('0')} copies of D'''E'''F'''G'''H'''I'''J'''K''' and {csh('53210')} copies of ABCDEFGHIJKLMNOPQRSTUV (number 3).")
  for i in range(width):
    for j in range(height):
      x, y = i*20, j*20
      index = pixels[i*width+j]
      if not index == 7:
        idx.append(index)
        for k in composition[index]:
          if k == '3':
            c3 +=1
            knorm = 'the three'
            cc = c3
          if k == 'T':
            cT +=1
            cc = cT
            knorm = 'the triangle'
          if k in 'Aaå@':
            if 'Aaå@'.index(k)==0:
              cA +=1
              cc = cA
              knorm = 'the A'
            if 'Aaå@'.index(k)==1:
              cAf +=1
              cc = cAf
              knorm = '''the flipped A (D'E'F'G'H'I'J'K')'''
            if 'Aaå@'.index(k)==2:
              cAa +=1
              cc = cAa
              knorm = """the rotated A (D''E''F''G''H''I''J''K'')"""
            if 'Aaå@'.index(k)==3:
              cAb +=1
              cc = cAb
              knorm = '''the rotated A (D\'\''E'\''F'\''G'\''H'\''I''\'J'\''K''\')'''
          if cc == 1:
            cth = 'first'
          elif cc == 2:
            cth = 'second'
          elif cc == 3:
            cth = 'third'
          elif cc == 4:
            cth = 'fourth'
          elif cc == 11:
            cth = '11th'
          elif cc == 12:
            cth = '12th'
          elif cc == 13:
            cth = '13th'
          elif str(cc)[-1] == '1':
            cth = str(cc)+'st'
          elif str(cc)[-1] == '2':
            cth = str(cc)+'nd'
          elif str(cc)[-1] == '3':
            cth = str(cc)+'rd'
          else:
            cth = str(cc)+'th'
          f.write(f"\n{str(step)}. Shift the {cth} copy of {knorm} right {round(i*0.08,5)} and up {round(j*0.08,5)}.")
          step+=1
      output.paste(tiles[index], (x, y), tiles[index])
  f.write(f"\n{step+1}. Reflect image over the y-axis.")

new_image = Image.new("RGBA", output.size, "WHITE")
new_image.paste(output, (0, 0), output) 
new_image.convert('RGB').save("output.png")
print(j*-0.08,i*0.08)
```
